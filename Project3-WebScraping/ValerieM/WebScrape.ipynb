```python
from bs4 import BeautifulSoup
import mechanize
import string

def isAscii(s):
    for c in s:
        if c not in string.ascii_letters:
            return False
    return True

url = "http://web2.nyrrc.org/cgi-bin/start.cgi/mar-programs/archive/archive_search.html"

browser = mechanize.Browser()
browser.addheaders = [
    ('user-agent', 'Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.9.2.3) Gecko/20100423 Ubuntu/10.04 (lucid) Firefox/3.6.3'),
    ('accept', 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8')
]
browser.set_handle_robots(False)

# get years
browser.open(url)
browser.form = list(browser.forms())[0]
select = browser.form.controls[3]
#years = [item.name for item in select.get_items()]
#for now, hard code years
years = ['1999','2000','2001','2002','2003','2004',
         '2005','2006','2007','2008','2009','2010',
         '2011','2013','2014','2015']
gender = ['M', 'F']

with open('output.txt', 'a') as f:
    
    f.write("Year FirstName LastName SexAge State Ctry Place GenderPlace AgePlace Time 10K 13.1M 20M MinPerMile AgePerf\n")
    # process year by year
    for year in years:
        for i in gender:
            print "Processing year=" + year + " Gender = " + i

            # open up search form again, fill out an appropriate year and submit
            browser.open(url)
            browser.form = list(browser.forms())[0]
            select = browser.form.controls[3]

            browser.form['input.searchyear'] = [year]
            browser.form['input.top'] = ['100']
            browser.form['top.type'] = [i]
            browser.form['search.method'] = ['search.top']
            browser.submit()

            soup = BeautifulSoup(browser.response())
            #        print soup
            for tr in soup.find_all('tr')[1:]:
                tds = tr.find_all('td')
                if (isAscii(tds[0].text)):
                    f.write("%s,%s," %
                        (year, tds[0].text))
                else:
                    f.write("%s,%s," %
                        (year, "JOHN"))    
                if tds[0].text == "ASTRID":
                    f.write("%s,%s," %
                           ("MILLER", tds[2].text))
                else:
                    if (isAscii(tds[1].text)):
                        f.write("%s,%s," %
                        (tds[1].text, tds[2].text))
                    else:
                        f.write("%s,%s," %
                        ("JOHN", tds[2].text))
                  
#                print 'tds[5] == ' + tds[5].text
                if tds[6].text != "United States":
                    f.write("%s," %
                       ("NA"))
                else:
                    f.write("%s," %
                       (tds[5].text))        
                f.write("%s,%s,%s,%s,%s,%s,%s,%s," %
                (tds[6].text, tds[8].text, tds[9].text,
                tds[10].text, tds[11].text, tds[13].text, 
                tds[14].text, tds[15].text ))
                if year < '2006':
                    f.write("%s,%s\n" %
                        (tds[16].text, tds[18].text))
                else:
                    f.write("%s,%s\n" %
                        (tds[22].text, tds[24].text))    
f.close()

##    print soup
##    print soup.find(text='Overall Winners').parent.parent.parent.find_next_sibling('tr').find('pre').text
```
